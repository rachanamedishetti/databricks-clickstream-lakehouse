```sql
CREATE VOLUME IF NOT EXISTS clickstream_prod.analytics.raw_events;
DROP TABLE IF EXISTS clickstream_prod.analytics.silver_clickstream;
DROP TABLE IF EXISTS clickstream_prod.analytics.bronze_clickstream;
DROP TABLE IF EXISTS clickstream_prod.analytics.gold_events_by_type;
DROP TABLE IF EXISTS clickstream_prod.analytics.gold_page_metrics;
DROP TABLE IF EXISTS clickstream_prod.analytics.silver_quarantine;
DROP TABLE IF EXISTS clickstream_prod.analytics.silver_user_dim;
DROP TABLE IF EXISTS clickstream_prod.analytics.gold_daily_active_users



SHOW TABLES IN clickstream_prod.analytics;

SELECT
  concat(
    'DROP ',
    table_type,
    ' IF EXISTS ',
    table_catalog, '.', table_schema, '.', table_name,
    ' CASCADE;'
  ) AS drop_stmt
FROM system.information_schema.tables
WHERE table_catalog = 'clickstream_prod'
  AND table_schema = 'analytics';

SELECT COUNT(*) AS bronze_count
FROM clickstream_prod.analytics.bronze_clickstream; --70

select count(*) from clickstream_prod.analytics.silver_clickstream; --58
select count(*) from clickstream_prod.analytics.silver_quarantine; --13
select count(*) from clickstream_prod.analytics.silver_user_dim; --57

select * from clickstream_prod.analytics.silver_clickstream;
select * from clickstream_prod.analytics.silver_quarantine;
select * from clickstream_prod.analytics.silver_user_dim where user_id = 'user_5';


select * from clickstream_prod.analytics.gold_daily_active_users;
select * from clickstream_prod.analytics.gold_events_by_type;
```
